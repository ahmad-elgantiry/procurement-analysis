```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading purchasing requisition table as a dataframe
purchase_requisition=pd.read_excel("List-of-Purchase-Requisition-1.xlsx","January 2018")
purchase_requisition.shape
```

```python
def cleanning_purchase_requisition(excel_file="List-of-Purchase-Requisition-1.xlsx",excel_sheet="January 2018"):
    #reading the data from excel file 
    purchase_requisition=pd.read_excel(excel_file,excel_sheet)
    # drop empty rows 
    purchase_requisition=purchase_requisition.dropna(how='all')
    # confirming that there is no duplicated PR_numbers ( it should be unique)
    duplicated_PR=purchase_requisition["PR_Number"].duplicated()
    duplicated_PR=purchase_requisition[duplicated_PR]["PR_Number"]
    # converting data types to correct datatypes
    purchase_requisition["PR Date"] = pd.to_datetime(purchase_requisition["PR Date"])
    purchase_requisition_convert_dict ={"PR_Number":object ,"ITEM Number" :object, "Store":object ,"QTY":int }
    purchase_requisition = purchase_requisition.astype(purchase_requisition_convert_dict)
    purchase_requisition=purchase_requisition.fillna(0)
    print(f"Number of found duplicated Purchasing requisition is  {len(duplicated_PR)} Purchasing Requisition")
    print(f"Duplicated_Purshasing Requisition found is {duplicated_PR.values}")
    print(purchase_requisition.dtypes)
    print(f"data shape is {purchase_requisition.shape}")
    purchase_requisition.to_csv("clean_purchase_requisition.csv",index=False)
    return purchase_requisition.head(2)
    

```

```python
cleanning_purchase_requisition()
```

```python
# reading Purchase Orders as a dataframe
purchasing_orders=pd.read_excel("List-of-Purchase-Orders.xlsx","January 2018")
purchasing_orders.head(2)
```

```python
def cleanning_purchase_orders(excel_file="List-of-Purchase-Orders.xlsx",excel_sheet="January 2018"):
    #reading the data from excel file 
    purchase_orders=pd.read_excel(excel_file,excel_sheet)
    # droping last row in Purchasing order because it contains only a total
    purchase_orders=purchase_orders.iloc[:-1]
    # drop empty rows 
    purchase_orders=purchase_orders.dropna(how='all')

    # confirming that there is no duplicated PO_numbers ( it should be unique)
    duplicated_PO=purchase_orders["PO_Number"].duplicated()
    duplicated_PO=purchase_orders[duplicated_PO]["PO_Number"]
    # converting data types to correct datatypes
    purchase_orders["PO_Date"] = pd.to_datetime(purchase_orders["PO_Date"])
    purchase_orders_convert_dict ={"PO_Number":object ,"ITEM Number" :object,"QTY":int,"PR_Number":object}
    purchase_orders = purchase_orders.astype(purchase_orders_convert_dict)
    purchase_orders=purchase_orders.fillna(0)
    print(f"Number of found duplicated PR_orders is  {len(duplicated_PO)} Purchasing Orders")
    print(f"Duplicated_Purshasing Orders found is {duplicated_PO.values}")
    print(purchase_orders.dtypes)
    print(f"data shape is {purchase_orders.shape}")
    purchase_orders.to_csv("clean_purchase_orders.csv",index=False)
    return purchase_orders.head(2)
```

```python
cleanning_purchase_orders()
```

```python
Good_Receipt=pd.read_excel("List-of-Goods-Receipt-Notes.xlsx","January 2018")
Good_Receipt.head(2)
```

```python
def cleanning_goods_receipt(excel_file="List-of-Goods-Receipt-Notes.xlsx",excel_sheet="January 2018"):
    #reading the data from excel file 
    Goods_Receipt=pd.read_excel(excel_file,excel_sheet)
    # drop empty rows 
    Goods_Receipt=Goods_Receipt.dropna(how='all')
    # confirming that there is no duplicated PO_numbers ( it should be unique)
    duplicated_PO=Goods_Receipt["PO_Number"].duplicated()
    duplicated_PO=Goods_Receipt[duplicated_PO]["PO_Number"]
    # converting data types to correct datatypes
    Goods_Receipt["Receipt Date"] = pd.to_datetime(Goods_Receipt["Receipt Date"])
    Goods_Receipt["Claimed Date"] = pd.to_datetime(Goods_Receipt["Claimed Date"])
    Goods_Receipt_convert_dict ={"PO_Number":object ,"ITEM Number" :str,"QTY Received":int}
    Goods_Receipt = Goods_Receipt.astype(Goods_Receipt_convert_dict)
    Goods_Receipt=Goods_Receipt.fillna(0)
    print(f"Number of found duplicated PO_orders is  {len(duplicated_PO)} Purchasing Orders")
    print(f"Duplicated_Purshasing Orders found is {duplicated_PO.values}")
    print(Goods_Receipt.dtypes)
    print(f"data shape is {Goods_Receipt.shape}")
    Goods_Receipt.to_csv("clean_Goods_receipt.csv",index=False)
    return Goods_Receipt.head(2)
```

```python
invoicess=pd.read_excel("List-of-Invoices-Received.xlsx","January 2018")
invoicess.head(2)
```

```python
def cleanning_invoices_table(excel_file="List-of-Invoices-Received.xlsx",excel_sheet="January 2018"):
    #reading the data from excel file 
    invoices=pd.read_excel(excel_file,excel_sheet)
    # drop empty rows 
    invoices=invoices.dropna(how='all')
    # confirming that there is no duplicated PO_numbers ( it should be unique)
    duplicated_PO=invoices["ORDER_NO"].duplicated()
    duplicated_PO=invoices[duplicated_PO]["ORDER_NO"]
    duplicated_invoices=invoices.duplicated().sum()
    # converting data types to correct datatypes
    invoices["INVDT_DATE"] = pd.to_datetime(invoices["INVDT_DATE"]) 
    invoices=invoices.rename(columns={"ORDER_NO": "PO_Number"}) 
    print(f"Number of found duplicated PO_orders is  {len(duplicated_PO)} Purchasing Orders")
    print(f"Duplicated_Purshasing Orders found is {duplicated_PO.values}")
    print(f"number of duplicated invoices is {duplicated_invoices}")
    print(invoices.dtypes)
    print(f"data shape is {invoices.shape}")
    invoices.to_csv("clean_invoices.csv",index=False)
    return invoices.head(2)
```

```python
cleanning_invoices_table()
```

```python
payments_table=pd.read_excel("Payments.xlsx","January 2018")
payments_table.head(2)
```

```python
def cleanning_payment_table(excel_file="Payments.xlsx",excel_sheet="January 2018"):
    #reading the data from excel file 
    payments=pd.read_excel(excel_file,excel_sheet)
    # drop empty rows 
    payments=payments.dropna(how='all')
    # confirming that there is no duplicated PO_numbers ( it should be unique)
    duplicated_invoice_number=payments["INVNO"].duplicated()
    duplicated_invoice_number=payments[duplicated_invoice_number]["INVNO"]
    duplicated_invoices=payments.duplicated().sum()
  
    print(f"Number of found duplicated invoice number  is  {len(duplicated_invoice_number)} invoices")
    print(f"Duplicated_invoice found is {duplicated_invoice_number.values}")
    print(f"number of duplicated invoices is {duplicated_invoices}")
    print(payments.dtypes)
    print(f"data shape is {payments.shape}")
    payments.to_csv("clean_payment_table.csv",index=False)
    return payments.head(2)
```

```python
cleanning_payment_table()
```

```python
def cleanning_vendors_table(excel_file="Vendor-File.xlsx"):
    #reading the data from excel file 
    vendors=pd.read_excel(excel_file)
    # drop empty rows if there is less than 3 values 
    vendors=vendors.dropna(thresh=3)
    # confirming that there is no duplicated suppliers code
    duplicated_supplier_code=vendors["SUPPLIER_CODE"].duplicated()
    duplicated_supplier_code=vendors[duplicated_supplier_code]["SUPPLIER_CODE"]
    print(f"Number of found duplicated supplier codes is  {len(duplicated_supplier_code)} vendor")
    print(f"Duplicated suppliers code found is {duplicated_supplier_code.values}")
    print(vendors.dtypes)
    print(f"data shape is {vendors.shape}")
    vendors.to_csv("clean_payment_table.csv",index=False)
    return vendors.head(2)
```
